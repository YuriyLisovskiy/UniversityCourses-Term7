```python
import math
import matplotlib.pyplot as plt
from numpy import arange


def create_plot(min_, max_, step, fn):
    # make it interactive
    plt.ion()
    plt.xlabel("The X axis")
    plt.ylabel("The Y axis")
    
    # Makes a 'log' or 'linear' scale
    plt.xscale('linear')
    xvalues = arange(min_, max_, step)
    plt.plot(xvalues, [fn(x) for x in xvalues], linestyle='--', label=fn.__name__)
    
    # plt.legend(loc="upper right")
    plt.legend(loc="best")


def sin_x(x):
    return math.sin(x)
```

```python
create_plot(-2*math.pi, 2*math.pi, 0.1, sin_x)
```

```python
class Witcher:
    
    def __init__(self, name, school, horse_name=None, age=None):
        self._name = name
        self._school = school
        self._horse_name = horse_name or 'Unknown'
        self._age = age or None
        
    def __str__(self):
        res = 'Witcher {} from school of the {}'.format(self.name, self.school)
        if self._age is not None:
            res += ' of age {}'.format(self.age)
        if self._horse_name is not 'Unknown':
            res += ' rides horse named {}'.format(self.horse_name)
        return res

    @property
    def name(self):
        return self._name
    
    @property
    def school(self):
        return self._school
    
    @property
    def horse_name(self):
        return self._horse_name
    
    @property
    def age(self):
        return self._age
    
    def fight(self):
        print('{} is fighting...'.format(self.name))

    
class TheWitcher4Game:
    
    def __init__(self, witchers):
        self._witchers = witchers
        
    def start(self):
        for witcher in self._witchers:
            witcher.fight()
        
    def studio(self):
        return 'CD PROJECT RED and KAROCH Studio'
        
    def finish(self):
        print('Designed by {}'.format(self.studio()))
            
    def __str__(self):
        return '{} Presents'.format(self.studio())
```

```python
geralt = Witcher(**{
    'name': 'Geralt',
    'school': 'Wolf',
    'horse_name': 'Roach',
    'age': 73
})

eskel = Witcher(
    name='Eskel',
    school='Wolf',
    horse_name='Scorpion',
    age=72
)

letho = Witcher('Letho', 'Snake')

game = TheWitcher4Game([geralt, eskel, letho])

print(game, end='\n\n')

game.start()

print()

game.finish()
```
